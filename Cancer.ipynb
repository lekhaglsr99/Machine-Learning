```python
import pandas as pd
```

```python
d=pd.read_csv("d:\cancer.txt")
print(d)
```

```python
d.info()
```

```python
d.replace('?',-999999,inplace=True)
```

```python
d.info()
```

```python
d.dropna(1,inplace=True)
```

```python
d.info()
```

```python
print(d)
```

```python
d.drop(['id'],1,inplace=True)
```

```python
import numpy as np
```

```python
x=np.array(d.drop(['class'],1))
y=np.array(d['class'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn import neighbors
```

```python
mdl=neighbors.KNeighborsClassifier()
```

```python
mdl.fit(x_train,y_train)
```

```python
acc=mdl.score(x_test,y_test)
```

```python
print(acc)
```

```python
val=[[4,2,1,1,1,2,3,2,1]]
pred=mdl.predict(val)
print(pred)
```
